```python
data_path = "repositories.json"
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_json(data_path)
```

```python
import matplotlib.pyplot as plt

x = []
y = []
z = []


for row in df.itertuples():
	x.append(row.LanguageCode)
	y.append(row.StargazersCount)
		

plt.bar(x, y, color = 'g', width = 0.72, label = "Stars")
plt.xlabel('LanguageCode')
plt.ylabel('Star Count')
plt.title('StarCount of Repos')
plt.legend()
plt.show()
```
